The survey was conducted during Level 4, with some retrospective questions relating to Pre-Lockdown (Feb 2020).

The first paper will be mainly descriptive and some simple associations.

The main thrust of the article will be the association between PA and health, in particular mental health (DASS-9) and physical health (WHO-5). So we would talk about that during periods of lockdown that people may not be able to exercise so what effect does that have on people’s health.
We might want to analyse the data by change in pre-post lockdown activity and their pre-post change in mental health (i.e. do people that maintain or increase PA maintain or increase mental and physical health).

We might then say well it’s the PA that does all of this, but to prove it control for PA and see if the DASS and WHO scores change. If they don’t, we can be pretty sure it’s to do with PA.

We may also say, well what activities were most effective, so can we correlate the change in PA with the type of activity the individuals were doing. So those who walked a dog, for example, were much better off than those that biked. This may help us to get some traction in saying that any exercise is good but in lockdown situations make sure you have a family dog or an online yoga subscription or whatever.

```r
# let's set up the tables for age, gender, prel4living, etc

age_groups <- mydata1 %>%
              count(Age_Group) %>%
              mutate(Age_pct = round( n * 100 / sum(n), 3) )
sex_tab <- mydata1 %>%
              count(Sex) %>%
              mutate(Sex_pct = round( n * 100 / sum(n), 3) )
pre_L4_living_arrangements <- mydata1 %>%
              count(Pre_Living) %>%
              mutate(pre4_pct = round( n * 100 / sum(n), 3) )
l4_living_arrangements <- mydata1 %>%
              count(L4_Living) %>%
              mutate(l4_pct = round( n * 100 / sum(n), 3) )
ethnicity <- mydata1 %>%
              count(Ethnicity_code) %>%
              mutate(eth_pct = round( n * 100 / sum(n), 3) )
comorbidity_affected_exercise <- mydata1 %>%
              count(CM_affect_exercise_YN) %>%
              mutate(c_a_e_pct = round( n * 100 / sum(n), 3) )  
pre_l4_exercise_behaviour <- mydata1 %>%
              count(Exercise_behaviour_pre) %>%
              mutate(e_b_p_pct = round( n * 100 / sum(n), 3) )
l4_exercise_behaviour <- mydata1 %>%
              count(L4_exercise_behaviour) %>%
              mutate(l_e_b_pct = round( n * 100 / sum(n), 3) )
ipaq_class <- mydata1 %>%
              count(IPAQ_CLASSIFICATION) %>%
              mutate(ipaq_pct = round( n * 100 / sum(n), 3) )

guideline_compliance_pre_l4 <- mydata1 %>%
              count(Pre_PA_guidelines) %>%
              mutate(guide_pct = round( n * 100 / sum(n), 3) )
qualification <- mydata1 %>%
              count(Qualification) %>%
              mutate(qual_pct = round( n * 100 / sum(n), 3) )
money <- mydata1 %>%
              count(L4_money) %>%
              mutate(money_pct = round( n * 100 / sum(n), 3) )
essential_worker_status <- mydata1 %>%
              count(Essential_Worker) %>%
              mutate(e_w = round( n * 100 / sum(n), 3) )
```

```r
## Let's build the conceptual models
## the first set of models are with WHO5
## Model 1 (WHO5 binarised with physical activity with mets alone)
whobin_met <- glm(whobin ~ met_thirds, family = binomial(link=logit), data = mydata3)
tidy(whobin_met) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

# whobin = 1 good mental health status, 0 = poor mental health status as per WHO5 scale
# shows that compared with lowest levels of exercise, those who were moderate 1500 - 2870, were 2.20 times as likely to have better 
# mental health status, and those with 2870- 6000, were about 3.42

# model 2, adjust for age

whobin_met1 <- glm(whobin ~ met_thirds + factor(Age_Group), family = binomial(link=logit), data = mydata3)
tidy(whobin_met1) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## similar for mets, after adjusting for age, and older people had better who5 mental health status

# model 3, adjust for age, sex
whobin_met2 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex), family = binomial(link=logit), data = mydata3)
tidy(whobin_met2) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## Men had slightly better mental health score as per WHO5, but not statistically significant,
## Age and mets hold their values

# model 4, adjust for age, comorbidity

whobin_met3 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN, family = binomial(link=logit), data = mydata3)
tidy(whobin_met3) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## those whose comorbid conditions limited their exercise abilities, nevertheless had better mental health status, 
# even after adjusting for all otehr variables, and the effects of mets remained unchanged
# including if their comorbid conditions did not limit their exercising abilities, would have a higher level of mental 
# health measured by WHO5 scale. This was after adjustment of all other variables!

#  model 5, adjust for age, comorbidity, and sitting

whobin_met4 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN + sit_thirds,
                  family = binomial(link=logit), data = mydata3)
tidy(whobin_met4) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## model 6, adjust for age, comorbidity, sitting, and ipaq

whobin_met5 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN + sit_thirds + factor(IPAQ_CLASSIFICATION),
                  family = binomial(link=logit), data = mydata3)
tidy(whobin_met5) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## this model suggests that the effect of mets is confounded by ipaq classification of physical activity and that ipaq classificaiton of 
# physical activity is a stronger predictor of mental health status than mets information alone

## model 7, adjust for age, comorbidity, sitting, and exercise behaviour in l4
whobin_met6 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN + sit_thirds + l4_exercise,
                  family = binomial(link=logit), data = mydata3)
tidy(whobin_met6) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## As with ipaq classification, this too is a strong predictor of mental health, and suggests that compared with those who were mostly
# sedentary, those who expressed a higher level of intent of exercising or moved "up" the ladder of intention or stages of change model
# also had higher scores in mental health as measured in WHO5

## model 8, adjust for age, comorbidity, sitting, and wealth
whobin_met7 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN + sit_thirds + factor(L4_money),
                  family = binomial(link=logit), data = mydata3)
tidy(whobin_met7) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## People who had more money had higher WHO5 score of satisfaction but this did not confound the association between mets or
# intensity of exercise and better mental health status
# we'd have to combine level 4 or we would have to discount level 5 of L4_money

## So the final model is model 9 for WHO5 where we wil adjust for age, comorbidity, sitting, L4_exercise behaviour and wealth

whobin_met8 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN + sit_thirds + l4_exercise + factor(L4_money),
                  family = binomial(link=logit), data = mydata3)
tidy(whobin_met8) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## We will divide the population into essential workers and non-essential workers and we will examine the performance of
# the two groups

essential_workers <- filter(mydata3, Essential_Worker == 1) # separate essential worker as a separate data set
not_essential_worker <- filter(mydata3, Essential_Worker != 1) # not essential worker

## Model 10 is same as model 9 but only for essential workers
whobin_met9 <- glm(whobin ~ met_thirds + factor(Age_Group) + factor(Sex) + CM_affect_exercise_YN + sit_thirds + l4_exercise + factor(L4_money),
                  family = binomial(link=logit), data = essential_workers)
tidy(whobin_met9) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## For essential workers the intention to exercise more was a more potent predictor than actual mets

## Model 11 is same as model 9 but only for non-essential workers

whobin_met10 <- glm(whobin ~ met_thirds + factor(Age_Group) + CM_affect_exercise_YN + sit_thirds + l4_exercise + factor(L4_money),
                  family = binomial(link=logit), data = not_essential_worker)
tidy(whobin_met10) %>%
mutate(odds_ratio = exp(estimate)) %>%
kable()

## for the non essential workers, the relationship between mets and WHO5 mental health issues were very similar to the general 
# non-segregated members of the public. 
```
